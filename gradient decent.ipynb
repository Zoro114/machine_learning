```python
import numpy as np
```

```python
load=np.genfromtxt("data.csv",delimiter =',',skip_header=1)
load
```

```python
def step_gradient(points,alpha,m,c):
    m_slope=0
    c_slope=0
    n=len(points)
    for i in range(n):
        x=load[i,0]
        y=load[i,1]
        m_slope+=(-2/n)*(y-m*x+c)*x
        c_slope+=(-2/n)*(y-m*x+c)   
        new_m=m-alpha*m_slope
        new_c=c-alpha*c_slope  
        return new_m,new_c
```

```python
def gd(points,alpha,num_iterations):
    m=0
    c=0
    for i in range(num_iterations):
        m, c=step_gradient(points,alpha,m,c)
        print(i," cost: ",cost(points,m,c))
    return m,c
```

```python
def cost(points,m,c):
    total_cost=0
    N=len(points)
    for i in range(N):
        x=points[i,0]
        y=points[i,1]
        total_cost+=(1/N)*((y-m*x-c)**2)
    return total_cost
```

```python
alpha=0.01
num_iterations=2500
m=gd(load,alpha,num_iterations)
print(m[0],m[1])
```
